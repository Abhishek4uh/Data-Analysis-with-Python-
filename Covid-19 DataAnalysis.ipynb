```python
import numpy as np
import pandas as pd
df=pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df
```

```python
df_1=df.iloc[:-1,:8]
df_2=df.iloc[:-1,9:]
fr=[df_1,df_2]
df_new=pd.concat(fr,axis=1)
```

```python
df_new
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states=np.arange(len(df_new['state_name']))
```

```python
total_states
```

```python
df_new['active']
```

```python
max(df['positive'])
```

## Total Positive Cases Based on States In India

```python

from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df_new['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(total_states, df_new['state_name'])  
plt.xlim(1,max(df_new['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

#### Active Number of Cases based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df_new['active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df_new['state_name'])  
plt.xlim(1,max(df_new['active'])+10) 
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df_new['new_active'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df_new['state_name'])  
plt.xlim(1,max(df_new['new_active'])+10) 
plt.xlabel(' New Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states,df_new['death'], align='center', alpha=0.5,  
                 color=(0.8,0.7,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(total_states, df_new['state_name'])  
plt.xlim(1,max(df_new['death'])+10) 
plt.xlabel('No of Death Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```
